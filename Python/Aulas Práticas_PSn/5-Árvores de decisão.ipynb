-  a) Crie uma função que dada uma lista de probabilidades de classes calcule o valor de impuridade
correspondente de acordo com o critério de Gini. Teste a sua função com o vetor
[0.25,0.25,0.25,0.25] que corresponde a quatro classes equiprováveis e cujo resultado deverá
ser 0.75.

```python
def Gini(probs):
    res=1
    for p in probs:
        res=res-p*p
    return res

Gini([0.25,0.25,0.25,0.25])
```

- b) Repita a alínea anterior cirando agora uma função para o critério da Entropia. O resultado com
o mesmo vetor de teste deverá ser 2.

```python
from math import log2

def Entropy(probs):
    res=0
    for p in probs:
        res=res-p*log2(p)
    return res
        
Entropy([0.25,0.25,0.25,0.25])
```

9 – Experimente agora a biblioteca sklearn que permite gerar uma árvore de decisão de uma forma
muito simples. Para tal terá que transformar todas as variáveis discretas em variáveis numéricas para
o que se recomenda a utilização de um dicionário do tipo:

- tt={'male':0, 'female':1, 'own':0, 'free':1, 'rent':2, 'little':0, 'quite rich':1, 'rich':2, 'moderate':1, 'radio/TV':0, 'Other':1, 'car':2}


Para gerar a árvore de decisão do sklearn basta fazer:
- from sklearn.tree import DecisionTreeClassifier as dtClass
- from sklearn.model_selection import cross_val_score
- …
- clf=dtClass(random_state=0)
- clf.fit(GCdata, GCTarget) # GCdata – predictors GCTarget – class


```python
from sklearn.tree import DecisionTreeClassifier as dtClass
from sklearn.model_selection import cross_val_score

tt={'male':0, 'female':1, 'own':0, 'free':1, 'rent':2, 'little':0, 'quite rich':1, 
    'rich':2, 'moderate':1, 'radio/TV':0, 'Other':1, 'car':2}

def transf(v):
    if v in tt:
        return(tt[v])
    return(v)

for c in data.columns:
    data[c]=data[c].apply(lambda x:transf(x))
    
GCdata=data.drop('Purpose',1)
GCTarget=data['Purpose']

clf=dtClass(random_state=0)
clf.fit(GCdata, GCTarget) # GCdata – predictors GCTarget – class
```
