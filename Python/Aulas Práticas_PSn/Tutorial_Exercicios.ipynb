Exercises (2)

• Take your program to to guess a number between 1 to 9 and add a
counter to tell the user “You guessed the number at the 5th attempt”

• Tune your message to say “1st”, “2nd”, “3rd” and “4th”

```python
import random

x=random.randint(1,9)
found=0
count=1
while found==0:
    guess=input("Guess a number between 1 and 9")
    if int(guess)==x:
        print("Correct! Congratulations")
        found=1
    else:
        print("Wrong number! Try again...")
        count+=1

if count==1:
    print("You guessed at the 1st attempt")
if count==2:
    print("You guessed at the 2nd attempt")
if count==3:
    print("You guessed at the 3rd attempt")    
else:
    print("You guessed at the "+str(count)+"th attempt")
```

Let’s make an animal classifier!
34

• Exercise:

    • Create a new script: animal_classifier.py
    • Give the user information: This classifier will
    identify your animal if it is either a dog, cat, fish,
    bird, human or duck
    • You will need user input for the following
    information:
    • Does the animal breathe air?
    • Does the animal have wings?
    • How many legs does it have?
    • Does it bark?
    • Does it Quack?

```python
print("Let me guess your animal!")

breath=input("does it breathe air? (Y/N) ")

if breath=='n' or breath=='N':
    print("Your animal is a fish!")
else:
    wings=input("does it have wings? (Y/N) ")
    if wings=='y' or wings=='Y':
        quack=input("does it quack? (Y/N) ")
        if quack=='y' or quack=='Y':
            print("Your animal is a Duck!")
        else:
            print("Your animal is a Bird!")
    else:
        legs=input("How many legs does it have? (2 or 4) ")
        if legs=='2':
            print("Your animal is an Human!")
        else:
            bark=input("Does it bark? (Y/N)")
            if bark=='y' or bark=='Y':
                print("Your animal is a Dog!")
            else:
                print("Your animal is a Cat")
            
```
