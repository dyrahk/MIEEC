```python
def cleanStr2float(s):

    ini=0
    last=len(s)-1

    if s[0]=='€':
        ini=1
    if s[last]=='M':
        return float(s[ini:last])*1000000
    if s[last]=='K':
        return float(s[ini:last])*1000
    return float(s[ini:last+1])

def cleanHeight(s):
   
    ini=0
    after=2
    last=len(s)
    if len(s)==3:
        return float(s[ini])*12*2.54+float(s[after])*2.54
    if len(s)==4:
        return float(s[ini])*12*2.54+float(s[after:last])*2.54

WorkPath = 'C:/Users/pedro/Desktop/Faculdade/5_2sem/PSn'

Data=pd.read_csv(WorkPath+'/FIFA.csv',converters={
    'Value':cleanStr2float,
    'Wage':cleanStr2float,
    'Height':cleanHeight
})
Data.drop(['ID','Photo','Flag','Club Logo','Special','International Reputation','Body Type','Real Face','Jersey Number'],axis=1,inplace=True)
```

```python
def chro_map(WHAT):

    url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
    state_geo = f'{url}/world-countries.json'
    state_unemployment = f'{url}/US_Unemployment_Oct2012.csv'
    state_data = pd.read_csv(state_unemployment)

    geolocator = Nominatim(user_agent="specify_your_app_name_here")
    countries=[]
    coord=[]
    wages_plus=[]
    state_data=[]

    i=0

    for country in Data3['Nationality']:
        if country not in countries:
            countries.append(str(country))

    for ov in Data3[WHAT]:
        wages_plus.append(ov)

    wages=[0]*len(countries)
    numbers=[0]*len(countries)

    indexes=np.linspace(0,len(countries),1)

    for country in Data3['Nationality']:
        if country in countries:
            #print(Data2['Overall'])
            wages[countries.index(country)]=wages[countries.index(country)]+wages_plus[countries.index(country)]
            numbers[countries.index(country)]=numbers[countries.index(country)]+1



    for country in countries:
        wages[countries.index(country)]=wages[countries.index(country)]/numbers[countries.index(country)]

    countries[countries.index('Central African Rep.')]='Central African Republic'
    countries[countries.index('FYR Macedonia')]='Macedonia'
    countries[countries.index('Korea DPR')]='Democratic People\'s Republic of Korea'
    countries[countries.index('United States')]='United States of America'
    countries[countries.index('England')]='United Kingdom'
    countries[countries.index('Congo')]='Democratic Republic of the Congo'
    countries[countries.index('Republic of Ireland')]='Ireland'



    country_wage_Dict = dict(zip(countries, wages))

    #for country in countries:

       # state_data[country]=wages[countries.index(country)]

    data = {'Country': countries, 'Attribute': wages}
    wage_df=pd.DataFrame.from_dict(data)

    m = folium.Map(zoom_start=2)
    bins = list(wage_df['Attribute'].quantile([0, 0.25, 0.5, 0.75, 1]))

    folium.Choropleth(
        geo_data=state_geo,
        name='choropleth',
        data=wage_df,
        columns=['Country', 'Attribute'],
        key_on='feature.properties.name',
        fill_color='BuPu',
        fill_opacity=0.7,
        line_opacity=0.2,
        bins=bins,
        legend_name='Attribute'
    ).add_to(m)

    folium.LayerControl().add_to(m)
    m.save('plot_data{}.html'.format(WHAT))
```

```python
def prob_potential(NAME):
    PlayerList=[]
    Player=[]
    index=np.arange(len(Data['Name']))
    check=0
    for i in index:
        if NAME in Data['Name'][i]:
            if Data['Potential'][i] <= Data['Overall'][i]:
                
                Player.append(Data['Name'][i])
                Player.append('Already at full potential')
                Player.append(Data['Potential'][i])
                Player.append(Data['Overall'][i])
                Player.append(Data['Club'][i])
                Player.append(Data['Nationality'][i])
                Player.append(Data4['Tag'][i])
                PlayerList.append(Player.copy())
                Player.clear()
                check=1
            else:
                CLUB=Data['Club'][i]
                COUNTRY=Data['Nationality'][i]
                nat_num=0
                nat_pos=0
                club_num=0
                club_pos=0
                for j in index:
                    if Data['Age'][j] >= 26:
                        if Data['Club'][j] is CLUB:
                            club_num=club_num+1
                            if Data['Potential'][j] <= Data['Overall'][j]:
                                club_pos=club_pos+1
                        else:
                            if Data['Nationality'][j] is COUNTRY:
                                nat_num=nat_num+1
                                if Data['Potential'][j] <= Data['Overall'][j]:
                                    nat_pos=nat_pos+1
                                    
                prob_final=((club_pos+1)/(club_num+2)+(nat_pos+1)/(nat_num+2))/2

                Player.append(Data['Name'][i])
                Player.append(prob_final)
                Player.append(Data['Potential'][i])
                Player.append(Data['Overall'][i])
                Player.append(Data['Club'][i])
                Player.append(Data['Nationality'][i])
                Player.append(Data4['Tag'][i])
                PlayerList.append(Player.copy())
                Player.clear()
                check=1

    if check is not 1:
        print("Player ", NAME, "can't be found")
    
    return PlayerList
```

```python
def prob_potential_tag(NAME):
    index=np.arange(len(Data2['Name']))
    check=0
    for i in index:
        if NAME in Data2['Name'][i]:
            CLUB=Data2['Club'][i]
            COUNTRY=Data2['Nationality'][i]
            check=1
    if check is 1:
        nat_num=0
        nat_pos=0
        club_num=0
        club_pos=0
        for i in index:
            #print(['Club'][i])
            if Data2['Age'][i] >= 26:
                if Data2['Club'][i] is CLUB:          
                    club_num=club_num+1
                    if Data2['Potential'][i] <= Data2['Overall'][i]:
                        club_pos=club_pos+1
                else:
                    if Data2['Nationality'][i] is COUNTRY:
                        nat_num=nat_num+1
                        if Data2['Potential'][i] <= Data2['Overall'][i]:
                            nat_pos=nat_pos+1
    prob=(((club_pos+1)/(club_num+2)+(nat_pos+1)/(nat_num+2))/2)*100
    return prob
```

```python
def teamNeeds(CLUB):
    Clubtested = df3[df3.Club == CLUB].reset_index(drop = True)

    #Clubtested = df3[df3['Club'].isin([CLUB])]
    #Clubtested= df3[df3['Club'].isin(['Arsenal'])]
    index = np.arange(len(Clubtested['Name']))

    WellRounded = 0

    #counters
    cStriker = 0
    cLeftW = 0
    cRightW = 0
    cMidf = 0
    cRback = 0
    cLback = 0
    cCback = 0
    cGoalie = 0

    #Get players information from club in question
    for i in index:

        #nome = Clubtested['Name'][i]
        pos = Clubtested['Position'][i]
        #rat = Clubtested['Overall'][i]

        if pos in Striker:
            cStriker=cStriker+1
        elif pos in LeftW:
            cLeftW=cLeftW+1
        elif pos in RightW:
            cRightW=cRightW+1
        elif pos in Midf:
            cMidf=cMidf+1
        elif pos in Rback:
            cRback=cRback+1
        elif pos in Lback:
            cLback=cLback+1
        elif pos in Cback:
            cCback=cCback+1
        else:
            cGoalie=cGoalie+1

    #posiçoes em falta
    PosFalta=[]

    if cStriker < 3:
        PosFalta.append('Strikers')
    if cLeftW < 2:
        PosFalta.append('Left Winger')
    if cRightW < 2:
        PosFalta.append('Right Winger')
    if cMidf < 7:
        PosFalta.append('Midfielder')
    if cRback < 2:
        PosFalta.append('Right Defender')
    if cLback < 2:
        PosFalta.append('Left Defender')
    if cCback < 4:
        PosFalta.append('Center Defender')
    if cGoalie < 3:
        PosFalta.append('Goalkeeper')
        
    if PosFalta == []:
        PosFalta.append('Nothing')
        
    PosFalta.insert(0,'This team needs: ')


    
    #counter for best player in each position
    bST=0
    bLW=0
    bRW=0
    bMid=0
    bRB=0
    bLB=0
    bCB=0
    bGK=0

    bST_rat=0
    bLW_rat=0
    bRW_rat=0
    bMid_rat=0
    bRB_rat=0
    bLB_rat=0
    bCB_rat=0
    bGK_rat=0


    for j in index:

        nome = Clubtested['Name'][j]
        pos = Clubtested['Position'][j]
        rat = Clubtested['Overall'][j]

        if (pos in Striker) and bST==0:
            bST_rat=Clubtested['Overall'][j]
            bST=1
        elif (pos in LeftW) and bLW==0:
            bLW_rat=Clubtested['Overall'][j]
            bLW=1
        elif (pos in RightW) and bRW==0:
            bRW_rat=Clubtested['Overall'][j]
            bRW=1
        elif (pos in Midf) and bMid==0:
            bMid_rat=Clubtested['Overall'][j]
            bMid=1
        elif (pos in Rback) and bRB==0:
            bRB_rat=Clubtested['Overall'][j]
            bRB=1
        elif (pos in Lback) and bLB==0:
            bLB_rat=Clubtested['Overall'][j]
            bLB=1
        elif (pos in Cback) and bCB==0:
            bCB_rat=Clubtested['Overall'][j]
            bCB=1
        elif (pos in Goalie) and bGK==0:
            bGK_rat=Clubtested['Overall'][j]
            bGK=1
        else:
            continue

    array_rat=[bST_rat, bLW_rat, bRW_rat, bMid_rat, bRB_rat, bLB_rat, bCB_rat, bGK_rat]

    a=array_rat.index(min(array_rat))

    PosFalta.append('And should improve their')
    
    if a==0:
        PosFalta.append('Striker')
    if a==1:
        PosFalta.append('Left Winger')
    if a==2:
        PosFalta.append('Right Winger')
    if a==3:
        PosFalta.append('Midfielder')
    if a==4:
        PosFalta.append('Right Back')
    if a==5:
        PosFalta.append('Left Back')
    if a==6:
        PosFalta.append('Center Back')
    if a==7:
        PosFalta.append('Goalkeeper')

#print('Stricker :' ,cStriker, 'Left Wing: ',cLeftW, 'Right wing: ', cRightW, 'Midfielder: ' ,cMidf, 'Right back: ' ,cRback, 'Left Back: ' ,cLback, ' Center back: ', cCback, 'Goalkeeper: ' ,cGoalie)
    return PosFalta
```

```python
class GUI(wx.Frame):

    def __init__(self, *args, **kwargs):
        super(GUI, self).__init__(*args, **kwargs)
        self.frame_number = 1
        self.InitUI()

    def InitUI(self):

        menubar = wx.MenuBar()

        fileMenu = wx.Menu()

        maps = wx.Menu()
        ov_map=maps.Append(wx.ID_ANY, 'Overall')
        wa_map=maps.Append(wx.ID_ANY, 'Wage')
        va_map=maps.Append(wx.ID_ANY, 'Value')
        
        fileMenu.AppendMenu(wx.ID_ANY, 'Maps', maps)
        
        fileMenu.AppendSeparator()
        
        Potent=fileMenu.Append(wx.ID_ANY, 'Potential Probability')
        Taggy=fileMenu.Append(wx.ID_ANY, 'Tags')
        Leags=fileMenu.Append(wx.ID_ANY, 'Leagues')
        Needs=fileMenu.Append(wx.ID_ANY, 'Building Advice')
        Costs=fileMenu.Append(wx.ID_ANY, 'Cost Eval')
        
        
        self.P_Array=[]
        

        qmi = wx.MenuItem(fileMenu, wx.ID_EXIT, '&Quit\tCtrl+W')
        fileMenu.AppendItem(qmi)

        self.Bind(wx.EVT_MENU, self.OnQuit, qmi)
        
        self.Bind(wx.EVT_MENU, self.OverallMap,ov_map)
        self.Bind(wx.EVT_MENU, self.WageMap,wa_map)
        self.Bind(wx.EVT_MENU, self.ValueMap,va_map)
        self.Bind(wx.EVT_MENU, self.PotProb,Potent)
        self.Bind(wx.EVT_MENU, self.LeagueShow,Leags)
        self.Bind(wx.EVT_MENU, self.Needed,Needs)
        self.Bind(wx.EVT_MENU, self.Cost,Costs)
        self.Bind(wx.EVT_MENU, self.Tag,Taggy)

        menubar.Append(fileMenu, '&File')
        self.SetMenuBar(menubar)

        self.SetSize((350, 250))
        self.SetTitle('PSn')
        self.Centre()
        
        panel = wx.Panel(self)  
        
        self.label1=wx.StaticText(panel, -1, "Search by Player Name:")
        self.label2=wx.StaticText(panel, -1, "       Overall:")
        self.label3=wx.StaticText(panel, -1, "Nationality:")
        self.label4=wx.StaticText(panel, -1, "           Club:")
        self.label5=wx.StaticText(panel, -1, "           Tags:")
        
        my_sizer = wx.BoxSizer(wx.VERTICAL)
        
        over_sizer=wx.BoxSizer(wx.HORIZONTAL)
        nat_sizer=wx.BoxSizer(wx.HORIZONTAL)
        tag_sizer=wx.BoxSizer(wx.HORIZONTAL)
        club_sizer=wx.BoxSizer(wx.HORIZONTAL)
        
        self.text_ctrl1 = wx.TextCtrl(panel)
        
        self.combobox=wx.ComboBox(panel)
        self.Bind(wx.EVT_COMBOBOX, self.onCombo)
        
        self.text_ctrl2 = wx.TextCtrl(panel)
        self.text_ctrl2.SetEditable(False)
        
        self.text_ctrl3 = wx.TextCtrl(panel)
        self.text_ctrl3.SetEditable(False)
        
        self.text_ctrl4 = wx.TextCtrl(panel)
        self.text_ctrl4.SetEditable(False)
        
        self.text_ctrl5 = wx.TextCtrl(panel)
        self.text_ctrl5.SetEditable(False)
        
        my_sizer.Add(self.label1, 0, wx.LEFT)
        my_sizer.Add(self.text_ctrl1, 0, wx.ALL | wx.EXPAND, 5)   
        
        my_btn = wx.Button(panel, label='Search')
        my_btn.Bind(wx.EVT_BUTTON, self.player_search)
        my_sizer.Add(my_btn, 0, wx.ALL | wx.CENTER, 5)  
        my_sizer.Add(self.combobox, 0, wx.ALL | wx.EXPAND, 5)
        
        
        
        over_sizer.Add(self.label2, 0, wx.ALL)
        over_sizer.Add(self.text_ctrl2, 0, wx.ALL | wx.EXPAND)
        
        nat_sizer.Add(self.label3, 0, wx.ALL)
        nat_sizer.Add(self.text_ctrl3, 0, wx.ALL | wx.EXPAND)
        
        club_sizer.Add(self.label4, 0, wx.ALL)
        club_sizer.Add(self.text_ctrl4, 0, wx.ALL | wx.EXPAND)
        
        tag_sizer.Add(self.label5, 0, wx.ALL)
        tag_sizer.Add(self.text_ctrl5, 0, wx.ALL | wx.EXPAND)
        
        my_sizer.Add(over_sizer, 0, wx.ALL)
        my_sizer.Add(nat_sizer, 0, wx.ALL)
        my_sizer.Add(club_sizer, 0, wx.ALL)
        my_sizer.Add(tag_sizer, 0, wx.ALL)
        
        panel.SetSizer(my_sizer)
        

    def OnQuit(self, e):
        self.Close()
    
    def WageMap(self, e):
        webbrowser.open_new_tab('plot_dataWage.html')
        
    def OverallMap(self, e):
        webbrowser.open_new_tab('plot_dataOverall.html')
        
    def ValueMap(self, e):
        webbrowser.open_new_tab('plot_dataValue.html')

    def PotProb(self, event):
        title = 'SubFrame {}'.format(self.frame_number)
        frame = ProbGridFrame(title=title)
        self.frame_number += 1
        
    def LeagueShow(self, event):
        title = 'SubFrame {}'.format(self.frame_number)
        frame = LeagueFrame(title=title)
        self.frame_number += 1
        
    def Needed(self, event):
        title = 'SubFrame {}'.format(self.frame_number)
        frame = NeededFrame(title=title)
        self.frame_number += 1
        
    def Cost(self, event):
        title = 'SubFrame {}'.format(self.frame_number)
        frame = CostFrame(title=title)
        self.frame_number += 1
        
    def Tag(self, event):
        title = 'SubFrame {}'.format(self.frame_number)
        frame = TagGridFrame(title=title)
        self.frame_number += 1

    def player_search(self, event):
        self.Sed_Players=P_Search(self.text_ctrl1.GetValue())
        for i in range(len(self.Sed_Players)):
            self.P_Array.append(self.Sed_Players[i][0])
        self.combobox.SetItems(self.P_Array.copy())
        self.P_Array.clear()

    def onCombo(self,event):
        comboValue= self.combobox.GetValue()
        for i in range(len(self.Sed_Players)):
            if comboValue == self.Sed_Players[i][0]:
                self.text_ctrl2.SetValue(str(self.Sed_Players[i][1]))
                self.text_ctrl3.SetValue(str(self.Sed_Players[i][3]))
                self.text_ctrl4.SetValue(str(self.Sed_Players[i][2]))
                self.text_ctrl5.SetValue(str(self.Sed_Players[i][4]))

def main():

    app = wx.App()
    ex = GUI(None)
    ex.Show()
    app.MainLoop()


if __name__ == '__main__':
    main()
```
