```python
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import numpy as np

ind1 = ctrl.Antecedent(np.arange(-2, 11, 1), 'Indice 1')
ind2 = ctrl.Antecedent(np.arange(-2, 11, 1), 'Indice 2')
ind3 = ctrl.Antecedent(np.arange(0, 9, 1), 'Indice 3')
print(ind1.universe)
print(ind2.universe)
print(ind3.universe)

gasInj=ctrl.Consequent(np.arange(-10,51,1), 'Gasolina Injetada')
print(gasInj.universe)
```

```python
ind1['VL'] = fuzz.trimf(ind1.universe, [-2, 0, 2])
ind1['L'] = fuzz.trimf(ind1.universe, [0, 2, 4])
ind1['M'] = fuzz.trimf(ind1.universe, [2, 5, 8])
ind1['H'] = fuzz.trimf(ind1.universe, [6, 8, 10])

ind1.view()
```

```python
ind2['L'] = fuzz.trimf(ind2.universe, [-2, 0, 2])
ind2['M'] = fuzz.trimf(ind2.universe, [0, 2, 4])
ind2['H'] = fuzz.trimf(ind2.universe, [2, 6, 10])

ind2.view()
```

```python
ind3['L'] = fuzz.trimf(ind3.universe, [0, 2, 4])
ind3['M'] = fuzz.trimf(ind3.universe, [2, 4, 6])
ind3['H'] = fuzz.trimf(ind3.universe, [4, 6, 8])

ind3.view()
```

```python
gasInj['Pouca'] = fuzz.trimf(gasInj.universe, [-10, 0, 20])
gasInj['Média'] = fuzz.trapmf(gasInj.universe, [0, 10, 20, 30])
gasInj['Muita'] = fuzz.trimf(gasInj.universe, [10, 30, 50])
gasInj.view()
```

```python
rule1 = ctrl.Rule((ind1['VL'] | ind1['L']) & (ind2['L'] | ind2['M']) & (ind3['L'] | ind3['M']),gasInj['Pouca'])
rule2 = ctrl.Rule((ind2['M'] | ind2['H']) & (ind3['M'] | ind3['H']),gasInj['Média'])
rule3 = ctrl.Rule((ind1['H']) & (ind2['L'] | ind2['H']) & (ind3['M']),gasInj['Muita'])
rule4 = ctrl.Rule((ind1['L'] | ind1['M']) & (ind2['L'] | ind2['M']) & (ind3['L'] | ind3['M']),gasInj['Média'])
```

```python
gasInj_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4])
gasInj_simulador = ctrl.ControlSystemSimulation(gasInj_ctrl)
```

```python
gasInj_simulador.input['Indice 1'] = 1
gasInj_simulador.input['Indice 2'] = 3
gasInj_simulador.input['Indice 3'] = 5
```

```python
gasInj_simulador.compute()
```

```python
print(gasInj_simulador.output['Gasolina Injetada'])
```

```python
ind1.view(sim=gasInj_simulador)
ind2.view(sim=gasInj_simulador)
ind3.view(sim=gasInj_simulador)
gasInj.view(sim=gasInj_simulador)
```
